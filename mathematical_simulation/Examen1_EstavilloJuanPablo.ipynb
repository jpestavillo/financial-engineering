```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import odeint

import matplotlib as mpl 
```

```python
a=1
b=1
def incisoa (variables,t):
        x1,x2=variables
        return [a*x2, -b*x1]
inicial=[5,5] #estos son datos del problema


tiempo=np.arange(0,50,.01)

resultado = odeint(incisoa,inicial,tiempo)
#odeint para integrar 
xx,yy=resultado.T


label_size= 16
mpl.rcParams['xtick.labelsize']= label_size
mpl.rcParams['xtick.labelsize']= label_size 


plt.plot(tiempo,xx,c= 'k', label="el bato")
plt.plot(tiempo,yy,c='m',label="la chava bipolar")
plt.legend(loc='best',prop={'size':16})
plt.xlabel('',fontsize=14)
plt.show()


```
